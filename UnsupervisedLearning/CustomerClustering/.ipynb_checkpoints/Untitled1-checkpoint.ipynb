```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import cluster, datasets, mixture
from sklearn.neighbors import kneighbors_graph
from sklearn.preprocessing import StandardScaler
from itertools import cycle, islice
```

```python
df = pd.read_csv('Mall_Customers.csv')
print(df)
```

```python
print(df['Age'].where(df['Gender'] == 'Male').dropna())
```

```python
fig = plt.figure(figsize=[8, 4])
ax = fig.add_subplot()

ax.set_xlabel('Age')
ax.set_ylabel('Spending Score')

ax.scatter(df['Age'].where(df['Gender'] == 'Male').dropna(), 
            df['Spending Score (1-100)'].where(df['Gender'] == 'Male').dropna(), c='blue')
ax.scatter(df['Age'].where(df['Gender'] == 'Female').dropna(), 
            df['Spending Score (1-100)'].where(df['Gender'] == 'Female').dropna(), c='red')

plt.show()
```

```python
fig = plt.figure(figsize=[8, 4])
ax = fig.add_subplot()

ax.set_xlabel('Annual Income in Thousands')
ax.set_ylabel('Spending Score')

ax.scatter(df['Annual Income (k$)'].where(df['Gender'] == 'Male').dropna(), 
            df['Spending Score (1-100)'].where(df['Gender'] == 'Male').dropna(), c='blue')

ax.scatter(df['Annual Income (k$)'].where(df['Gender'] == 'Female').dropna(), 
            df['Spending Score (1-100)'].where(df['Gender'] == 'Female').dropna(), c='red')

plt.show()
```

```python
fig = plt.figure(figsize=[8, 4])
ax = fig.add_subplot()

ax.set_xlabel('Annual Income in Thousands')
ax.set_ylabel('Spending Score')

ages = [20, 40, 60, 80]
cols = ['red', 'orange', 'blue', 'black']

for i, col in zip(ages, cols):

    ax.scatter(df['Annual Income (k$)'].where((df['Age'] <= i) & (df['Age'] > i-20)).dropna(), 
        df['Spending Score (1-100)'].where((df['Age'] <= i) & (df['Age'] > i-20)).dropna(), 
        s=30, c=col)


plt.show()
```

```python

fig = plt.figure(figsize=[16, 16])
ax = fig.add_subplot(projection='3d')

ax.set_xlabel('Age')
ax.set_ylabel('Income')
ax.set_zlabel('Spending Score')

ax.scatter3D(df['Annual Income (k$)'], 
             df['Age'], 
             df['Spending Score (1-100)'], 
             c='red', s=50);

```

```python
np.random.seed(0)
X = StandardScaler().fit_transform(df[['Spending Score (1-100)', 'Annual Income (k$)']])

bandwidth = cluster.estimate_bandwidth(X, quantile=0.15, n_samples=len(X))
meanshift = cluster.MeanShift(bandwidth)
meanshift.fit(X)
```

```python
fig = plt.figure(figsize=[8, 8])
ax = fig.add_subplot()

labels_unique = np.unique(meanshift.labels_)
n_clusters = len(labels_unique)
print(n_clusters)
cols = ['red', 'orange', 'blue', 'violet', 'black']
ax.set_title("Meanshift")
for i in range(n_clusters):
    my_members = meanshift.labels_ == i
    ax.scatter(X[my_members, 0], X[my_members, 1], 
               s=100, c=cols[i], edgecolors='white')
```
